## 1. Load and Explore TFRecord Data

```python
# ─── Load TFRecord Data ────────────────────────────────────────────────

print("[EDA] Loading TFRecord data...")
print("\n📊 Loading splits: train, eval, test\n")

# Load all splits
X_train, y_train = load_tfrecord_data("train")
X_eval, y_eval = load_tfrecord_data("eval")
X_test, y_test = load_tfrecord_data("test")

print(f"✅ Data loaded successfully!")
print(f"\n{'='*60}")
print(f"DATASET OVERVIEW")
print(f"{'='*60}")

# Summary table
summary_data = {
    'Split': ['Train', 'Eval', 'Test', 'Total'],
    'Samples': [len(X_train), len(X_eval), len(X_test), len(X_train) + len(X_eval) + len(X_test)],
    'Shape': [f'{X_train.shape}', f'{X_eval.shape}', f'{X_test.shape}', '-'],
    'Fire Pixels': [
        int((y_train == 1).sum()),
        int((y_eval == 1).sum()),
        int((y_test == 1).sum()),
        int((y_train == 1).sum() + (y_eval == 1).sum() + (y_test == 1).sum())
    ],
}

df_summary = pd.DataFrame(summary_data)
print("\n" + df_summary.to_string(index=False))

# Fire ratio
total_fire = (y_train == 1).sum() + (y_eval == 1).sum() + (y_test == 1).sum()
total_pixels = X_train.size / 12 + X_eval.size / 12 + X_test.size / 12
fire_ratio = total_fire / total_pixels * 100

print(f"\n📊 Class Distribution:")
print(f"   Fire pixels: {int(total_fire):,} ({fire_ratio:.2f}%)")
print(f"   No-fire pixels: {int(total_pixels - total_fire):,} ({100-fire_ratio:.2f}%)")
print(f"   Imbalance ratio: {(1-fire_ratio)/fire_ratio:.2f}x")
print(f"\n{'='*60}")
```

## 2. Fire vs No-Fire Class Distribution

```python
# ─── Class Distribution Analysis ───────────────────────────────────────

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Fire pixel counts per sample
fire_counts_train = (y_train == 1).sum(axis=(1, 2))
fire_counts_eval = (y_eval == 1).sum(axis=(1, 2))
fire_counts_test = (y_test == 1).sum(axis=(1, 2))

# Plot 1: Distribution of fire pixels per sample (train)
axes[0, 0].hist(fire_counts_train, bins=50, color='#FF6B6B', alpha=0.7, edgecolor='white')
axes[0, 0].set_title('Train: Fire Pixels per Sample', fontweight='bold', fontsize=12)
axes[0, 0].set_xlabel('Fire Pixel Count')
axes[0, 0].set_ylabel('Sample Count')
axes[0, 0].axvline(fire_counts_train.mean(), color='yellow', linestyle='--', linewidth=2, label=f'Mean: {fire_counts_train.mean():.0f}')
axes[0, 0].legend()

# Plot 2: Eval + Test
axes[0, 1].hist(fire_counts_eval, bins=50, color='#4ECDC4', alpha=0.7, label='Eval', edgecolor='white')
axes[0, 1].hist(fire_counts_test, bins=50, color='#45B7D1', alpha=0.5, label='Test', edgecolor='white')
axes[0, 1].set_title('Eval & Test: Fire Pixels per Sample', fontweight='bold', fontsize=12)
axes[0, 1].set_xlabel('Fire Pixel Count')
axes[0, 1].set_ylabel('Sample Count')
axes[0, 1].legend()

# Plot 3: Pie chart of samples with/without fire
train_with_fire = (fire_counts_train > 0).sum()
train_without_fire = (fire_counts_train == 0).sum()
axes[1, 0].pie([train_with_fire, train_without_fire], 
              labels=[f'Has Fire\n({train_with_fire})', f'No Fire\n({train_without_fire})'],
              colors=['#FF6B6B', '#2ECC40'], autopct='%1.1f%%',
              startangle=90, textprops={'fontsize': 11})
axes[1, 0].set_title('Train: Samples with/without Fire', fontweight='bold', fontsize=12)

# Plot 4: Fire pixel counts across all splits
data_for_box = [fire_counts_train, fire_counts_eval, fire_counts_test]
bp = axes[1, 1].boxplot(data_for_box, labels=['Train', 'Eval', 'Test'], patch_artist=True)
for patch, color in zip(bp['boxes'], ['#FF6B6B', '#4ECDC4', '#45B7D1']):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1, 1].set_title('Fire Pixels Distribution (Box Plot)', fontweight='bold', fontsize=12)
axes[1, 1].set_ylabel('Fire Pixel Count')
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_01_class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n📊 Class Statistics:")
print(f"   Train:  {train_with_fire:,} samples with fire, {train_without_fire:,} without")
print(f"   Eval:   {(fire_counts_eval > 0).sum():,} samples with fire, {(fire_counts_eval == 0).sum():,} without")
print(f"   Test:   {(fire_counts_test > 0).sum():,} samples with fire, {(fire_counts_test == 0).sum():,} without")
```

## 3. Visualize Satellite Patches with Fire

```python
# ─── Visualize Sample Satellite Patches ────────────────────────────────

# Select samples with high fire content
fire_counts = (y_test == 1).sum(axis=(1, 2))
high_fire_idx = np.argsort(fire_counts)[-6:]  # Top 6 with most fire

fig, axes = plt.subplots(6, 4, figsize=(16, 18))

for row, idx in enumerate(high_fire_idx):
    patch = X_test[idx]
    label = y_test[idx]
    
    # Channel 0: NDVI
    im0 = axes[row, 0].imshow(patch[0], cmap='RdYlGn', vmin=-2000, vmax=10000)
    axes[row, 0].set_title(f'NDVI (Ch 0)', fontsize=9, fontweight='bold')
    axes[row, 0].axis('off')
    
    # Channel 1: Elevation
    im1 = axes[row, 1].imshow(patch[1], cmap='terrain', vmin=0, vmax=3000)
    axes[row, 1].set_title(f'Elevation (Ch 1)', fontsize=9, fontweight='bold')
    axes[row, 1].axis('off')
    
    # Channel 5: Max Temperature
    im2 = axes[row, 2].imshow(patch[5], cmap='hot', vmin=270, vmax=330)
    axes[row, 2].set_title(f'Max Temp (Ch 5)', fontsize=9, fontweight='bold')
    axes[row, 2].axis('off')
    
    # Channel 11: Previous Fire + Label
    im3 = axes[row, 3].imshow(label, cmap='Reds', vmin=0, vmax=1)
    fire_px = (label == 1).sum()
    axes[row, 3].set_title(f'Fire Label: {int(fire_px)} px', fontsize=9, fontweight='bold', color='#FF6B6B')
    axes[row, 3].axis('off')

plt.suptitle('High-Fire Samples: Key Satellite Channels', fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_02_sample_patches.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\n📸 Visualized {len(high_fire_idx)} samples with highest fire content")
```

## 4. Channel Statistics Analysis (12 Landsat-8 Features)

```python
# ─── Channel Statistics ────────────────────────────────────────────────

# Compute statistics for each channel
stats_data = []

for ch_idx, feat_name in enumerate(INPUT_FEATURES):
    train_ch = X_train[:, ch_idx, :, :]
    eval_ch = X_eval[:, ch_idx, :, :]
    
    stats_data.append({
        'Channel': ch_idx,
        'Feature': feat_name,
        'Full Name': FEATURE_INFO[feat_name]['full_name'],
        'Mean': train_ch.mean(),
        'Std': train_ch.std(),
        'Min': train_ch.min(),
        'Max': train_ch.max(),
        'Unit': FEATURE_INFO[feat_name]['unit'],
    })

df_stats = pd.DataFrame(stats_data)

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Mean values for all channels
ax = axes[0, 0]
ax.barh(range(len(df_stats)), df_stats['Mean'], color='#4ECDC4', edgecolor='white')
ax.set_yticks(range(len(df_stats)))
ax.set_yticklabels(df_stats['Feature'])
ax.set_xlabel('Mean Value')
ax.set_title('Mean Value per Channel (Train Set)', fontweight='bold', fontsize=12)
ax.grid(axis='x', alpha=0.3)

# Plot 2: Std values
ax = axes[0, 1]
ax.barh(range(len(df_stats)), df_stats['Std'], color='#FF6B6B', edgecolor='white')
ax.set_yticks(range(len(df_stats)))
ax.set_yticklabels(df_stats['Feature'])
ax.set_xlabel('Standard Deviation')
ax.set_title('Std Dev per Channel (Train Set)', fontweight='bold', fontsize=12)
ax.grid(axis='x', alpha=0.3)

# Plot 3: Value ranges (min-max)
ax = axes[1, 0]
for i, row in df_stats.iterrows():
    ax.barh(i, row['Max'] - row['Min'], left=row['Min'], height=0.6, 
           color=plt.cm.viridis(i / len(df_stats)), edgecolor='white', linewidth=0.5)
ax.set_yticks(range(len(df_stats)))
ax.set_yticklabels(df_stats['Feature'])
ax.set_xlabel('Value Range')
ax.set_title('Min-Max Range per Channel', fontweight='bold', fontsize=12)
ax.grid(axis='x', alpha=0.3)

# Plot 4: Distribution of a few key channels
ax = axes[1, 1]
key_channels = [0, 1, 5, 11]  # NDVI, elevation, temp, prev fire
for ch_idx in key_channels:
    data = X_train[:, ch_idx, :, :].flatten()
    ax.hist(data, bins=50, alpha=0.5, label=INPUT_FEATURES[ch_idx], density=True)
ax.set_xlabel('Pixel Value')
ax.set_ylabel('Density')
ax.set_title('Distribution of Key Channels', fontweight='bold', fontsize=12)
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_03_channel_statistics.png', dpi=150, bbox_inches='tight')
plt.show()

# Print detailed statistics table
print("\n📊 Channel Statistics (Training Set):")
print("=" * 120)
print(df_stats.to_string(index=False))
print("=" * 120)
```

## 5. Climate & Environmental Features Correlation

```python
# ─── Feature Correlation with Fire ───────────────────────────────────────

# Sample data for correlation (use test set)
n_sample = min(500, len(X_test))
X_sample = X_test[:n_sample].reshape(n_sample, 12, -1).mean(axis=2)  # Mean across spatial dims
y_sample = ((y_test[:n_sample] == 1).sum(axis=(1, 2)) > 0).astype(int)  # Binary: has fire or not

# Create correlation dataframe
corr_data = pd.DataFrame(X_sample, columns=INPUT_FEATURES)
corr_data['has_fire'] = y_sample

# Compute correlations with fire
correlations = corr_data.corr()['has_fire'].drop('has_fire').sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Correlation with fire
ax = axes[0]
colors_corr = ['#FF6B6B' if x > 0 else '#4ECDC4' for x in correlations.values]
ax.barh(range(len(correlations)), correlations.values, color=colors_corr, edgecolor='white')
ax.set_yticks(range(len(correlations)))
ax.set_yticklabels(correlations.index)
ax.set_xlabel('Correlation with Fire')
ax.set_title('Feature Correlation with Fire Presence', fontweight='bold', fontsize=12)
ax.axvline(0, color='white', linestyle='-', linewidth=0.8)
ax.grid(axis='x', alpha=0.3)

# Plot 2: Fire presence vs key features (scatter)
ax = axes[1]
key_indices = [0, 5, 9, 11]  # NDVI, max temp, ERC, prev fire
for i, ch_idx in enumerate(key_indices):
    feature_values = X_sample[:, ch_idx]
    noise = np.random.normal(0, 0.02, len(feature_values))
    y_jitter = y_sample + noise
    ax.scatter(feature_values, y_jitter, alpha=0.3, s=20, 
              label=INPUT_FEATURES[ch_idx])

ax.set_xlabel('Feature Value (normalized)')
ax.set_ylabel('Fire Presence (jittered)')
ax.set_title('Key Features vs Fire Presence', fontweight='bold', fontsize=12)
ax.set_ylim(-0.2, 1.2)
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_04_feature_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n📈 Feature Correlation with Fire (samples with fire vs without):")
for feat, corr in correlations.items():
    print(f"   {feat:15s}: {corr:+.4f}")
```

## 7. Preprocessing Requirements & Normalization Ranges

```python
# ─── Preprocessing Requirements & Normalization ────────────────────────

print("\n" + "="*80)
print("PREPROCESSING REQUIREMENTS FOR MODEL TRAINING")
print("="*80)

normalization_params = []

for ch_idx, feat_name in enumerate(INPUT_FEATURES):
    train_ch = X_train[:, ch_idx, :, :].flatten()
    
    # Compute normalization statistics
    mean = train_ch.mean()
    std = train_ch.std()
    min_val = train_ch.min()
    max_val = train_ch.max()
    
    # Check for outliers (values > 3 std from mean)
    outlier_count = ((np.abs(train_ch - mean) > 3 * std).sum())
    outlier_pct = outlier_count / len(train_ch) * 100
    
    normalization_params.append({
        'Channel': ch_idx,
        'Feature': feat_name,
        'Mean': mean,
        'Std': std,
        'Min': min_val,
        'Max': max_val,
        'Outliers (%)': outlier_pct,
    })

norm_df = pd.DataFrame(normalization_params)

print("\n📊 Per-Channel Normalization Parameters:")
print("="*80)
print(norm_df.to_string(index=False))
print("="*80)

# Generate Python code for model preprocessing
print("\n🔧 Recommended Preprocessing Code (for model):")
print("""
# StandardScaler normalization per channel
from sklearn.preprocessing import StandardScaler

# Per-channel means and stds from training data
CHANNEL_MEANS = [""" + ", ".join([f"{x:.2f}" for x in norm_df['Mean'].values]) + """]
CHANNEL_STDS = [""" + ", ".join([f"{x:.2f}" for x in norm_df['Std'].values]) + """]

def preprocess_patch(x):
    '''Normalize satellite patch using training statistics.'''
    x = x.astype(np.float32)
    for ch in range(12):
        x[ch] = (x[ch] - CHANNEL_MEANS[ch]) / (CHANNEL_STDS[ch] + 1e-6)
    return x
""")

print("\n⚠️ Outlier Detection:")
high_outlier = norm_df[norm_df['Outliers (%)'] > 5]
if len(high_outlier) > 0:
    print("   Channels with >5% outliers (consider clipping or log-transform):")
    for _, row in high_outlier.iterrows():
        print(f"   - {row['Feature']:15s}: {row['Outliers (%)']:.2f}%")
else:
    print("   ✅ No channels with excessive outliers")
```

## 8. Train/Eval/Test Split Quality Check

```python
# ─── Split Quality Verification ────────────────────────────────────────

# Compare distributions across splits to detect data leakage or imbalance
splits_data = {
    'Train': (X_train, y_train),
    'Eval': (X_eval, y_eval),
    'Test': (X_test, y_test),
}

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Compute statistics for each split
split_stats = []

for idx, (split_name, (X_split, y_split)) in enumerate(splits_data.items()):
    fire_counts = (y_split == 1).sum(axis=(1, 2))
    
    # Plot distribution
    axes[idx].hist(fire_counts, bins=40, alpha=0.7, 
                   color=['#FF6B6B', '#4ECDC4', '#45B7D1'][idx], edgecolor='white')
    axes[idx].set_title(f'{split_name}: Fire Pixel Distribution', fontweight='bold', fontsize=12)
    axes[idx].set_xlabel('Fire Pixels per Sample')
    axes[idx].set_ylabel('Sample Count')
    axes[idx].axvline(fire_counts.mean(), color='yellow', linestyle='--', linewidth=2, 
                      label=f'Mean: {fire_counts.mean():.1f}')
    axes[idx].legend()
    axes[idx].grid(axis='y', alpha=0.3)
    
    # Collect statistics
    split_stats.append({
        'Split': split_name,
        'Samples': len(X_split),
        'Fire Pixels (Total)': int((y_split == 1).sum()),
        'Mean Fire/Sample': fire_counts.mean(),
        'Std Fire/Sample': fire_counts.std(),
        'Min Fire/Sample': fire_counts.min(),
        'Max Fire/Sample': fire_counts.max(),
        '% Samples with Fire': (fire_counts > 0).sum() / len(X_split) * 100,
    })

plt.suptitle('Data Split Quality Check', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_07_split_quality.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✅ Split Consistency Analysis:")
split_df = pd.DataFrame(split_stats)
print(split_df.to_string(index=False))

# Check for anomalies
print("\n🔍 Quality Checks:")
for stat_col in ['Mean Fire/Sample', 'Std Fire/Sample', '% Samples with Fire']:
    values = split_df[stat_col].values
    if len(values) > 1 and np.std(values) / np.mean(values) > 0.2:
        print(f"   ⚠️ {stat_col}: High variance across splits (potential leakage or imbalance)")
    else:
        print(f"   ✅ {stat_col}: Consistent across splits")
```

## 9. Channel Interaction Patterns

```python
# ─── Multi-Channel Interaction Patterns ────────────────────────────────

# Key channel indices: NDVI(0), Elevation(1), Temp(5), Prev Fire(11)
key_channels = {0: 'NDVI', 1: 'Elevation', 5: 'Max Temp', 11: 'Prev Fire'}

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

# Sample test data
n_interact = min(1000, len(X_test))
X_interact = X_test[:n_interact]
y_interact = y_test[:n_interact]

# Create binary fire label (has fire or not)
fire_label = ((y_interact == 1).sum(axis=(1, 2)) > 0).astype(int)

# Plot interactions between key channels
plot_idx = 0

for (ch1_id, ch1_name), (ch2_id, ch2_name) in [
    ((0, 'NDVI'), (5, 'Max Temp')),
    ((0, 'NDVI'), (1, 'Elevation')),
    ((5, 'Max Temp'), (1, 'Elevation')),
    ((0, 'NDVI'), (11, 'Prev Fire')),
    ((5, 'Max Temp'), (11, 'Prev Fire')),
]:
    # Get spatial mean for each channel
    ch1_vals = X_interact[:, ch1_id, :, :].reshape(len(X_interact), -1).mean(axis=1)
    ch2_vals = X_interact[:, ch2_id, :, :].reshape(len(X_interact), -1).mean(axis=1)
    
    # Plot with fire/no-fire distinction
    fire_mask = fire_label == 1
    axes[plot_idx].scatter(ch1_vals[~fire_mask], ch2_vals[~fire_mask], 
                          alpha=0.3, s=30, color='#2ECC40', label='No Fire')
    axes[plot_idx].scatter(ch1_vals[fire_mask], ch2_vals[fire_mask], 
                          alpha=0.5, s=50, color='#FF6B6B', label='Has Fire', edgecolor='white', linewidth=0.5)
    
    axes[plot_idx].set_xlabel(ch1_name, fontweight='bold')
    axes[plot_idx].set_ylabel(ch2_name, fontweight='bold')
    axes[plot_idx].set_title(f'{ch1_name} vs {ch2_name}', fontweight='bold')
    axes[plot_idx].legend()
    axes[plot_idx].grid(alpha=0.3)
    
    plot_idx += 1

# Remove last empty subplot
fig.delaxes(axes[-1])

plt.suptitle('Channel Interaction Patterns: Fire vs No-Fire', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_06_channel_interactions.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n🔍 Key Insight: Fire samples tend to cluster in specific channel combinations")
```

## 10. Fire Pixel Clustering Analysis

```python
# ─── Fire Pixel Clustering & Spatial Distribution ────────────────────

from scipy import ndimage

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

clustering_stats = []

for split_name, X_split, y_split in [('Train', X_train, y_train), 
                                       ('Eval', X_eval, y_eval), 
                                       ('Test', X_test, y_test)]:
    # Find samples with fire
    fire_samples_idx = np.where((y_split == 1).sum(axis=(1, 2)) > 0)[0]
    
    if len(fire_samples_idx) == 0:
        continue
    
    # Analyze clustering for each fire sample
    num_clusters_list = []
    cluster_size_list = []
    fire_pixel_counts = []
    
    for idx in fire_samples_idx[:100]:  # Sample 100 for speed
        label = y_split[idx].astype(int)
        
        # Count fire pixels
        fire_count = (label == 1).sum()
        if fire_count == 0:
            continue
        
        fire_pixel_counts.append(fire_count)
        
        # Label connected components (clusters)
        labeled_array, num_features = ndimage.label(label)
        num_clusters_list.append(num_features)
        
        # Get cluster sizes
        for cluster_id in range(1, num_features + 1):
            cluster_size = (labeled_array == cluster_id).sum()
            cluster_size_list.append(cluster_size)
    
    if len(num_clusters_list) == 0:
        continue
    
    # Plot 1: Number of clusters per sample
    row = 0 if split_name == 'Train' else (1 if split_name == 'Eval' else 0)
    col = 0 if split_name == 'Train' else (1 if split_name == 'Eval' else 2)
    
    if split_name == 'Train':
        axes[0, 0].hist(num_clusters_list, bins=30, color='#FF6B6B', alpha=0.7, edgecolor='white')
        axes[0, 0].set_title(f'Train: Fire Clusters per Sample', fontweight='bold')
        axes[0, 0].set_xlabel('Number of Clusters')
        axes[0, 0].set_ylabel('Sample Count')
        
        axes[1, 0].hist(cluster_size_list, bins=30, color='#FF6B6B', alpha=0.7, edgecolor='white')
        axes[1, 0].set_title(f'Train: Cluster Sizes', fontweight='bold')
        axes[1, 0].set_xlabel('Pixels per Cluster')
        axes[1, 0].set_ylabel('Frequency')
        
    elif split_name == 'Eval':
        axes[0, 1].hist(num_clusters_list, bins=30, color='#4ECDC4', alpha=0.7, edgecolor='white')
        axes[0, 1].set_title(f'Eval: Fire Clusters per Sample', fontweight='bold')
        axes[0, 1].set_xlabel('Number of Clusters')
        axes[0, 1].set_ylabel('Sample Count')
        
        axes[1, 1].hist(cluster_size_list, bins=30, color='#4ECDC4', alpha=0.7, edgecolor='white')
        axes[1, 1].set_title(f'Eval: Cluster Sizes', fontweight='bold')
        axes[1, 1].set_xlabel('Pixels per Cluster')
        axes[1, 1].set_ylabel('Frequency')
        
    else:  # Test
        axes[0, 2].hist(num_clusters_list, bins=30, color='#45B7D1', alpha=0.7, edgecolor='white')
        axes[0, 2].set_title(f'Test: Fire Clusters per Sample', fontweight='bold')
        axes[0, 2].set_xlabel('Number of Clusters')
        axes[0, 2].set_ylabel('Sample Count')
        
        axes[1, 2].hist(cluster_size_list, bins=30, color='#45B7D1', alpha=0.7, edgecolor='white')
        axes[1, 2].set_title(f'Test: Cluster Sizes', fontweight='bold')
        axes[1, 2].set_xlabel('Pixels per Cluster')
        axes[1, 2].set_ylabel('Frequency')
    
    clustering_stats.append({
        'Split': split_name,
        'Avg Clusters': np.mean(num_clusters_list),
        'Max Clusters': np.max(num_clusters_list),
        'Avg Cluster Size': np.mean(cluster_size_list) if cluster_size_list else 0,
        'Avg Fire Pixels': np.mean(fire_pixel_counts),
    })

plt.suptitle('Fire Pixel Clustering Analysis', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'eda_05_clustering_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n🔥 Clustering Statistics (Fire Samples):")
clustering_df = pd.DataFrame(clustering_stats)
print(clustering_df.to_string(index=False))
```

## 11. Interactive Model Prediction — Upload & Predict
Upload a `.npy` or `.npz` satellite patch (shape `12×64×64`) to visualize and predict fire.

```python
# --- Load Trained Model & Normalization Stats ---
import torch
import json as _json

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load normalization stats (saved during training)
info_path = PROJECT_ROOT / 'checkpoints' / 'data_info.json'
with open(info_path) as f:
    _data_info = _json.load(f)
MEANS = np.array(_data_info['means'], dtype=np.float32)
STDS  = np.array(_data_info['stds'],  dtype=np.float32)

# Load model
from src.models import get_model
model_pred, _ = get_model('unet_lite', in_channels=12)

ckpt_path = PROJECT_ROOT / 'checkpoints' / 'best_model.pth'
if not ckpt_path.exists():
    ckpt_path = PROJECT_ROOT / 'checkpoints' / 'final_model.pth'

ckpt = torch.load(str(ckpt_path), map_location=device, weights_only=False)
model_pred.load_state_dict(ckpt['model_state_dict'])
model_pred = model_pred.to(device).eval()

print(f'✅ Model loaded: {ckpt_path.name}')
print(f'   Device     : {device}')
print(f'   Parameters : {sum(p.numel() for p in model_pred.parameters()):,}')
print(f'   Norm stats : {len(MEANS)} channels loaded')
```

```python
# --- Upload Satellite Patch (.npy / .npz) ---
import io
import ipywidgets as widgets
from IPython.display import display, clear_output

uploaded_patch = None  # global: holds (12, 64, 64) array after upload

upload_btn = widgets.FileUpload(accept='.npy,.npz', multiple=False, description='Upload Patch')
status_out = widgets.Output()

def on_upload(change):
    global uploaded_patch
    with status_out:
        clear_output()
        val = upload_btn.value
        if not val:
            return

        # ipywidgets 8.x: val is a tuple of dicts with keys 'name', 'content'
        # ipywidgets 7.x: val is a dict keyed by filename
        if isinstance(val, dict):
            fname = list(val.keys())[0]
            raw   = val[fname]['content']
            name  = fname.lower()
        else:
            info  = val[0]
            name  = info['name'].lower()
            raw   = bytes(info['content'])

        if name.endswith('.npy'):
            patch = np.load(io.BytesIO(raw))
        elif name.endswith('.npz'):
            data = np.load(io.BytesIO(raw))
            keys = ['patch', 'data', 'X', 'x', 'input']
            key  = next((k for k in keys if k in data), list(data.keys())[0])
            patch = data[key]
        else:
            print('Unsupported format. Use .npy or .npz'); return

        # Handle (H, W, C) -> (C, H, W)
        if patch.ndim == 3 and patch.shape == (64, 64, 12):
            patch = patch.transpose(2, 0, 1)

        if patch.shape != (12, 64, 64):
            print(f'Wrong shape: {patch.shape}. Need (12, 64, 64)'); return

        uploaded_patch = patch.astype(np.float32)
        print(f'Loaded: {name}')
        print(f'   Shape : {uploaded_patch.shape}')
        print(f'   Range : [{uploaded_patch.min():.2f}, {uploaded_patch.max():.2f}]')
        print('   Run the cells below to visualize and predict.')

upload_btn.observe(on_upload, names='value')
display(widgets.VBox([upload_btn, status_out]))
```

```python
# --- Satellite Image: False-Colour RGB Composite ---
assert uploaded_patch is not None, '⚠️ Upload a patch first (run the cell above).'

def pnorm(arr, lo=2, hi=98):
    """Percentile normalise to [0, 1]."""
    lo_v, hi_v = np.percentile(arr, [lo, hi])
    return np.clip((arr - lo_v) / (hi_v - lo_v + 1e-8), 0, 1)

# False-colour: R = Max Temperature (ch 5), G = NDVI (ch 0), B = Elevation (ch 1)
rgb = np.stack([pnorm(uploaded_patch[5]),
                pnorm(uploaded_patch[0]),
                pnorm(uploaded_patch[1])], axis=-1)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(rgb, interpolation='nearest')
ax.set_title(
    'Satellite Patch — False-Colour Composite\n'
    'R = Max Temperature  |  G = NDVI  |  B = Elevation',
    fontsize=13, fontweight='bold'
)
ax.axis('off')
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'pred_satellite_composite.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved → outputs/pred_satellite_composite.png')
```

```python
# --- Fire Prediction ---
import matplotlib.colors as mcolors
from src.predict import predict_patch, generate_prediction_report

assert uploaded_patch is not None, '⚠️ Upload a patch first.'

prob_map, pred_mask = predict_patch(model_pred, uploaded_patch, MEANS, STDS, device)
report = generate_prediction_report(uploaded_patch, prob_map, pred_mask)

cmap_fire = mcolors.ListedColormap(['#1a1a2e', '#e94560'])
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Panel 1: NDVI input
axes[0].imshow(uploaded_patch[0], cmap='RdYlGn')
axes[0].set_title('NDVI (input)', fontweight='bold', fontsize=12)
axes[0].axis('off')

# Panel 2: Fire probability map
im = axes[1].imshow(prob_map, cmap='hot', vmin=0, vmax=1)
axes[1].set_title('Fire Probability Map', fontweight='bold', fontsize=12)
axes[1].axis('off')
plt.colorbar(im, ax=axes[1], fraction=0.046, pad=0.04)

# Panel 3: Binary prediction mask
axes[2].imshow(pred_mask, cmap=cmap_fire, vmin=0, vmax=1)
axes[2].set_title(f'Predicted Fire Mask  ({int(pred_mask.sum())} px)',
                  fontweight='bold', fontsize=12, color='#FF6B6B')
axes[2].axis('off')

plt.suptitle('🔥 Wildfire Prediction Results', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'pred_result.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\n📊 Prediction Report:')
print(f'   Risk Level    : {report["risk_level"]}')
print(f'   Fire Detected : {report["fire_detected"]}')
print(f'   Fire Pixels   : {report["fire_pixels"]} / {report["total_pixels"]}')
print(f'   Fire Coverage : {report["fire_percentage"]:.2f}%')
print(f'   Max Prob      : {report["max_probability"]:.4f}')
print(f'   Mean Prob     : {report["mean_probability"]:.4f}')
```
